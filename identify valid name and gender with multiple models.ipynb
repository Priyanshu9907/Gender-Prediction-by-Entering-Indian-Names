```python
def predict_name(name):
    name_vectorized = vectorizer.transform([name])
    predictions = []
    
    classifiers = [("Decision Tree ",classifierdt_name), 
                   ("Support Vector ",classifiersvc_name),
                   ("Random Forest ",classifierrf_name), 
                   ("Logistic Regression ",classifierlr_name),
                   ("KNN ",classifierknn_name)]
    
    for name,classifier in classifiers:
        prediction = classifier.predict(name_vectorized)
        if prediction[0] == 1:
            predictions.append(f"{name}: Valid Name")
        else: 
            predictions.append(f"{name}: Invalid Name")
    
    return "\n".join(predictions)
```

```python
def predict_name_most_occurance(name):
    name_vectorized = vectorizer.transform([name])
    predictions = []
    
    classifiers = [("Decision Tree", classifierdt_name), 
                   ("Support Vector", classifiersvc_name),
                   ("Random Forest", classifierrf_name), 
                   ("Logistic Regression", classifierlr_name), 
                   ("KNN", classifierknn_name)]
    
    for _, classifier in classifiers:
        prediction = classifier.predict(name_vectorized)
        predictions.append(prediction[0])
    
    valid_count = predictions.count(1)
    invalid_count = predictions.count(0)
    
    if valid_count > invalid_count:
        return "Valid Name"
    else:
        return "Invalid Name"
```
